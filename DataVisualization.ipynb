```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('fenerbahce_beko_cleaned.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.scatterplot(data=df , x="Points", y="Performance Index Rating", hue="IsHome")
plt.xlabel("Say")
plt.ylabel("Verimlilik")
plt.title("Say vs Verimlilik")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(data=df, x="Three-point %", y="Performance Index Rating", hue="IsHome")
plt.xlabel("lk Yzdesi")
plt.ylabel("Verimlilik")
plt.title("lk Yzdesi vs Verimlilik")
plt.show()
```

```python
sns.set_style("darkgrid")
sns.lmplot(data=df, x="Two-point %", y="Performance Index Rating", hue="IsHome")
plt.xlabel("kilik Yzdesi")
plt.ylabel("Verimlilik")
plt.title("kilik Yzdesi vs Verimlilik")
plt.show()
```

```python
sns.set_style("dark")
sns.lmplot(data=df, x="Free-throw %", y="Performance Index Rating", height=5)
plt.xlabel("Serbest At Yzdesi")
plt.ylabel("Verimlilik")
plt.title("Serbest At Yzdesi vs Verimlilik")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.kdeplot(df['Performance Index Rating'], fill=True, color='red')
plt.title('Verimlilik Puan Dalm')
plt.xlabel('Verimlilik')
plt.ylabel('Younluk')
plt.show()
```

```python
sns.set_style("white")
plt.figure(figsize=(13,6))

plt.subplot(1,3,1)
sns.boxplot(df["Blocks"])
plt.title("Blok")

plt.subplot(1,3,2)
sns.boxplot(df["Steals"])
plt.title("Top alma")

plt.subplot(1,3,3)
sns.boxplot(df["Turnovers"])
plt.title("Top Kayb")

plt.tight_layout()
plt.show()
```

```python
sns.countplot(x="IsHome", data=df, palette='pastel', hue="IsHome", legend=False)
plt.xticks([0,1], ["Away", "Home"])
plt.title("Ev Sahibi vs Deplasman")
plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Korelasyon Matrisi")
plt.show()
```

```python
corr = df.corr(numeric_only=True)[["Performance Index Rating"]].sort_values("Performance Index Rating", ascending=False)[1:]

plt.figure(figsize=(4,6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Verimlilik Puan')
plt.show()
```

```python
top10 = df['Opposing Team'].value_counts().head(10)

plt.figure(figsize=(8,8))
plt.pie(top10, labels=top10.index, autopct='%1.1f%%', startangle=140)
plt.title("FBB' nin En ok Karlat 10 Takm", loc="center")
plt.show()
```
